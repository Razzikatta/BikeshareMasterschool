```python
#creating a function for filtering the data
def get_filters():

#greeting the user
    print('Hello, Let us explore some of the US bikeshare data!')

#filter the city 
    while True:
        city = input("Which city would you like- Chicago, New York City, or Washington? ").lower()
        if city in CITY_DATA.keys():
            break
        else:
            print("This is not a valid city, please choose from the 3 options")
#filter the month
#list of the available month options, data is limited to January through June

    months = ['january', 'february', 'march', 'april', 'may', 'june', 'all']
    while True:
        month = input("Which month would you like to choose from between January and June? select one or all ").lower()
        if month in months:
            break
        else:
            print("This is not a valid month, please select again")
            
#filter the day
#list of weekday options
    days = ['sunday', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'all']
    while True:
        day = input("Which day would you like to look at? select one or all ").lower()
        if day in days:
            break
        else:
            print("This is not a valid day, please select again")
        
    print('`'*80)
    return city, month, day
```

```python
#function for loading the data out of the previous definition 
def load_data(city, month, day): 

    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """
        # load data from the csv file into a dataframe    
    df = pd.read_csv(CITY_DATA[city])
    
        # convert the Start Time column from object to datetime 
    df['Start Time'] = pd.to_datetime(df['Start Time'])
        
        # extract month and day from Start Time to create new columns
    df['month'] = df['Start Time'].dt.month_name()
    df['day'] = df['Start Time'].dt.day_name()
    
        # filter by month if applicable, create the new dataframe
    if month != 'all':
        df = df[df['month'] == month.title()]
        
        # filter by day of week to create the new dataframe
    if day != 'all':
        df = df[df['day'] == day.title()] 
        
        # filter by hour
    df['hour'] = df['Start Time'].dt.hour
    
    return df
```

```python
def time_stats(df):
      
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating the most popular travel times...\n')
    start_time = time.time()

    # TO DO: display the most common month
    popular_month = df['month'].mode()[0]
    print("the most common month is:", df['month'].mode()[0])
    
    # TO DO: display the most common day 
    popular_day = df['day'].mode()[0]
    print("the most common day is:", df['day'].mode()[0])

    # TO DO: display the most common start hour
    popular_hour = df['hour'].mode()[0]
    print("the most common start hour is:", df['hour'].mode()[0])
    
    print('\nThis took % seconds.' % (time.time() - start_time))
    
    print('-'*40)
```

```python
def station_stats(df):
    
    """Displays statistics on the most popular stations and trips."""
    
    print('\nCalculating the most popular station\n')
    start_time = time.time()
    
    # TO DO: display most commonly used start station
    popular_start_station = df['Start Station'].mode()[0]
    print('The most common start station is: ', popular_start_station)
    
    # TO DO: display most commonly used end station
    popular_end_station = df['End Station'].mode()[0]
    print('The most common end station is: ', popular_end_station)
    
    # TO DO: display most frequent combination of start station and end station trip
    start_end_combo = df['Start Station'].mode()[0] + " " + df['End Station'].mode()[0]
    
    print('The most popular start and end station combo is: \n', start_end_combo)

    return (popular_start_station, popular_end_station, start_end_combo)
    print(station_stats(df))
    
    print("\nThis took % seconds.") % (time.time() - start_time)
    print('`'*60)
```

```python
def trip_duration_stats(df):

    """Displays statistics on the total and average trip duration."""

    print('\nCalculating trip duration...\n')
    start_time = time.time()

    # TO DO: display total travel time
    total_travel_time = df['Trip Duration'].sum()
    print('total travel time is: ', total_travel_time)
    
    # TO DO: display mean travel time
    average_travel_time = df['Trip Duration'].mean()
    print('average travel time is: ', average_travel_time)
    
    print("\nThis took % seconds." % (time.time() - start_time))
    print("-" * 40)
```

```python
def user_stats(df):
        
    print('\nCalculating User Stats...\n')
    start_time = time.time()
    
    """Displays statistics on bikeshare users."""
          
    # TO DO: Display earliest, most recent, and most common year of birth, if column in dataframe 
    if 'Birth Year' in df.columns:
        years = df['Birth Year']          
        earliest_year = (years).min()
        most_recent_year = (years).max()
        most_common_year = (years).mode()
        print(f"Counts of years:\n {years} \n")
    else:
        print("There is no birth year column available for this city") 
                            
    # print value counts for each user type
    user_types = df['User Type'].value_counts()
    print(user_types)
                            
    # TO DO: Display counts of gender, if gender column is available in the dataframe selected 
                            
    if 'gender' in df.columns:                   
        gender_count = df["gender"].value_counts()
        #print("counts gender".gender_count)
        print(f"Counts of gender:\n {gender_count} \n")                    
    else:
        print("There is no gender column available for this city")                             

    print("\nThis took % seconds." % (time.time() - start_time))
    print('-'*80)                              
```

```python
#letting the user see 5 rows at a time out of the data they selected, until choosing to stop

def show_row_data(df):
    
    row = 0
    while True:
        view_row_data = input("Would you like to view the row data? reply yes or no.\n")
        if view_row_data == 'yes':
            print(df.iloc[row : row + 5])
            row += 5
        elif view_row_data == 'no':
            break
        else:
            print("Sorry, this is not a valid input, please choose yes or no")
```

```python
#main function definition
def main():
    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)

        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)
        show_row_data(df)

        restart = input('\nWould you like to restart? enter "y" for yes or "n" for no.\n')
        if restart.lower() != 'y':
            break

if __name__ == "__main__":
	main()
```
